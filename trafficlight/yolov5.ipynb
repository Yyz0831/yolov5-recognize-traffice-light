```python
#ÂáÜÁ°ÆÁéáËÆ°ÁÆó
import cv2
def rgb2hsv(r, g, b):
    r, g, b = r/255.0, g/255.0, b/255.0
    mx = max(r, g, b)
    mn = min(r, g, b)
    df = mx-mn
    if mx == mn:
        h = 0
    elif mx == r:
        h = (60 * ((g-b)/df) + 360) % 360
    elif mx == g:
        h = (60 * ((b-r)/df) + 120) % 360
    elif mx == b:
        h = (60 * ((r-g)/df) + 240) % 360
    if mx == 0:
        s = 0
    else:
        s = df/mx
    v = mx
    return h, s, v
def getcolor(path,i):
    img=cv2.imread("test/dayClip"+str(i)+"/frames/"+path[:-4]+".jpg")[:, :, ::-1]
    data = []
    red=0
    green=0
    yellow=0
    with open("yolov5/resulttxt/dayClip"+str(i)+"/"+path,"r") as f:
        for line in f.readlines():
            line = line.strip("\n")
            line = line.split()
            if line!=[]:
                line[0]=int(line[0])
                line[1]=int(line[1])
                line[2]=int(line[2])
                line[3]=int(line[3])
                data.append(line)
    for i in range(len(data)):
        x_len=data[i][2]-data[i][0]
        y_len=data[i][3]-data[i][1]
        if x_len>=y_len:
            maxl=x_len
            minl=y_len
            x_large=True
        else:
            maxl=y_len
            minl=x_len
            x_large=False
        #print(path,maxl,minl,maxl/minl)
        if (maxl>40)and(minl>20):
            normal=1
            for j in range(int(maxl/minl+normal)):
                if x_large==True:
                    x_location=data[i][0]+(maxl*(j+1))/(int(maxl/minl+normal)+1)
                    y_location=data[i][1]+minl/2
                else:
                    y_location=data[i][1]+(maxl*(j+1))/(int(maxl/minl+normal)+1)
                    x_location=data[i][0]+minl/2
                colorRGB=img[int(y_location)][int(x_location)]
                colorHSV=rgb2hsv(colorRGB[0],colorRGB[1],colorRGB[2])
                if (colorHSV[2]>0.25)and(colorHSV[1]>0.3):
                    if (colorHSV[0]<50)or(colorHSV[0]>340):
                        red=red+1
                    elif (colorHSV[0]>50)and(colorHSV[0]<65):
                        yellow=yellow+1
                    elif (colorHSV[0]<200)and(colorHSV[0]>65):
                        green=green+1
    if red>green:
        if red<yellow:
            result="yellow"
        else:
            result="red"
    elif red<green:
        if green<yellow:
            result="yellow"
        else:
            result="green"
    else:
        if yellow>green:
            result="yellow"
        else:
            result="green"
    return result
def pdt(s):
    if s=="go":
        return "green"
    elif s=="st":
        return "red"
    elif s=="wa":
        return"yellow"
```

```python
#ÁªüËÆ°
import os
total=0
right=0
for i in range(1,14):
    Ftotal=0
    Fright=0
    with open("test/annotation/dayClip"+str(i)+"/frameAnnotationsBULB.csv") as fd:
        path_list=os.listdir("yolov5/resulttxt/dayClip"+str(i))
        for p in path_list:
            result=getcolor(p,i)
            total=total+1
            Ftotal=Ftotal+1
            line=""
            if i<10:
                a=0
            else:
                a=1
            while line[12:(27+a)]!=p[:-4]:
                line=fd.readline()
            rpdt=pdt(line[(32+a):(34+a)])
            if result==rpdt:
                right=right+1
                Fright=Fright+1
    print(i,Ftotal,Fright,Fright/Ftotal)
print(total,right,right/total)
```

```python
import plotly.express as px
import numpy as np
import math
def hsv2rgb(h, s, v):
    h = float(h)
    s = float(s)
    v = float(v)
    h60 = h / 60.0
    h60f = math.floor(h60)
    hi = int(h60f) % 6
    f = h60 - h60f
    p = v * (1 - s)
    q = v * (1 - f * s)
    t = v * (1 - (1 - f) * s)
    r, g, b = 0, 0, 0
    if hi == 0: r, g, b = v, t, p
    elif hi == 1: r, g, b = q, v, p
    elif hi == 2: r, g, b = p, v, t
    elif hi == 3: r, g, b = p, q, v
    elif hi == 4: r, g, b = t, p, v
    elif hi == 5: r, g, b = v, p, q
    r, g, b = int(r * 255), int(g * 255), int(b * 255)
    return r, g, b
pic = np.zeros([1, 1, 3], dtype = 'uint8')
hsv=[179.6341463414634, 0.9162011173184357, 0.7019607843137254]
rgb=hsv2rgb(hsv[0],hsv[1],hsv[2])
pic[:,:,0]=245.5
pic[:,:,1]=rgb[1]
pic[:,:,2]=rgb[2]
print(pic)
fig = px.imshow(pic,
                zmin = 0,
                zmax = 255)
fig.show()
```
